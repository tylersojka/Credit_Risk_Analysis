```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import os
```

```python
# read in the dataset into a dataframe
salary_data = Path("../Resources/Salary_Data.csv")
df = pd.read_csv(salary_data)
df.head()
```

```python
# plot the data
plt.scatter(df.YearsExperience, df.Salary)
plt.xlabel('Years of Experience')
plt.ylabel('Salary in USD')
plt.show()
```

```python
# assign and format the data to X, to meet the requirements of the Scikit-learn library: The years of experience is the independent variable
X = df.YearsExperience.values.reshape(-1,1)
```

```python
# look at the first 5 entries of X
X[:5]
```

```python
# check the shape of the data, X is 30 samples with a single feature/column
X.shape
```

```python
# assign the target variable, Salary, to y. Y is the dependent variable because we assume that employee salary depends on the years of experience.
y = df.Salary
```

```python
# create an instance of the linear regression model from sklearn.linear_model's linearregression
model = LinearRegression()
```

```python
# fit the data to the model
model.fit(X, y)
```

```python
y_pred = model.predict(X)
print(y_pred.shape)
```

```python
plt.scatter(X, y)
plt.plot(X, y_pred, color='red')
plt.show()
```

```python
print(model.coef_)
print(model.intercept_)
```
